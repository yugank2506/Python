```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

##Import The Libraries And Dataset
```

```python
data=pd.read_csv('heart.csv') 
```

```python
##Get Information About Our Dataset Like Total Number Rows, Total Number of Columns, Datatypes of Each Column And Memory Requirement

data.info()
## tell all the data about mean, variance, other things 
data.describe()  
```

```python
##Check Null Values In The Dataset

data.isnull().sum()
```

```python
##Check For Duplicate Data and Drop Them
data.duplicated().any()
data.drop_duplicates()
```

```python
## using seaborn and matplotlib

sns.relplot(x='sex',y='thalach',data=data)
sns.relplot(x='chol',y='thalach',hue='sex',data=data)
sns.relplot(x='chol',y='thalach',hue='sex',style='slope',data=data)
sns.catplot(x='slope',y='trestbps',hue='sex',kind='strip', data=data)
```

```python
## printing heat map
a=data.corr()
sns.heatmap(a,annot=True)
```
